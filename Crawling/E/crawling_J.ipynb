```python
import numpy as np
import pandas as pd

import requests
from bs4 import BeautifulSoup 
```

```python
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.71 Safari/537.36'}
url = 'https://www.saramin.co.kr/zf_user/jobs/list/domestic'
res = requests.get(url, headers = headers)
res.raise_for_status()
soup = BeautifulSoup(res.text, 'lxml') 
```

```python
# 각각 채용공고 -> each_company
total_company = soup.find('div', attrs = {'class' : 'list_body'})
each_company = total_company.findAll('div', attrs = {'class' : 'list_item'})
```

```python
# 첫번째 a tag = 회사명
each_company[0].a.attrs['title']
```

```python
# 두번째 a tag = 제목 - job_tlt로 따오기
job_title = each_company[0].find('div', attrs = {'class' : 'job_tit'})
job_title.a.attrs['title']
```

```python
# 첫 40개
cnt = 0
df1 = pd.DataFrame(columns = ['company', 'job'])

for each in each_company:
    comp_name = each.a.attrs['title']
    job = each.find('div', attrs = {'class' : 'job_tit'})
    job_name = job.a.attrs['title']
    
    # 행 새로 추가
    cnt += 1
    df1.loc[cnt,:] = [comp_name, job_name]
    
    if cnt == 40:
        break
```
